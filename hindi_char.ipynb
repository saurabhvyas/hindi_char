```python
# this will create a dictionary oh hindi graphemes
# input : path to text file ( use cat prior to this file to generate a single text file )
# output : dictionary

# note this module requires uniseg python packages
# this will assume preprocessing has already been done 
```

```python
input_txt_file='/media/saurabh/New Volume3/feeny_dataset/modi_js/public/final_modi.txt'
```

```python
with open(input_txt_file, 'r') as myfile:
    data=myfile.read() #.replace('\n', '')
```

```python
graphemes={}
```

```python
from uniseg.graphemecluster import grapheme_clusters
```

```python
a = u"विश्व की मानवता को ललकारा है |2"
```

```python


graphemes_list=[]
```

```python
for i in grapheme_clusters(data):
    graphemes_list.append(i)
    
```

```python
len(graphemes_list)
```

```python
graphemes_list=list(set(graphemes_list))
graphemes_list
```

```python
for i,j in enumerate(graphemes_list):
    #print(i)
    graphemes[j]=i
```

```python
graphemes
```

```python


def sentence_to_grapheme(sentence):
    input_sentence_list=[]
    for i in grapheme_clusters(sentence):
    #input_sentence_list.append(i)
     # return dictionary value of i
        
        #print(i)
        
        input_sentence_list.append(graphemes[i])
    return input_sentence_list
    
```

```python
sentence_to_grapheme(a)
```

```python

char_black_list = "!@#$?.,|:()1234567890"

def remove_blacklisted_chars(input):
    for char in char_black_list:
        input = input.replace(char,"")
    return input
```
